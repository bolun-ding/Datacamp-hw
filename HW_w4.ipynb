4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
np.random.seed(0)
size = 100
x = np.random.rand(size)
y = x + np.random.normal(0, 0.3, size)
```

```python
df = pd.DataFrame({'Series1': x , 'Series2': y})
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(df_corr)
```

5. Now, create a scatter plot of two heatmap entries that appear to have negative correlation.

```python
np.random.seed(0)
size = 100
x = np.random.rand(size)
y = -0.8*x + np.random.normal(0, 0.2, size)
df = pd.DataFrame({'Series1': x , 'Series2': y})
df_corr = df.corr()
df_corr
sns.heatmap(df_corr)
```
